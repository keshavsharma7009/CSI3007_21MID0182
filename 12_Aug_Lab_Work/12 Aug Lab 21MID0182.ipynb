```python
class Empl:
    no_of_employees = 0
    raise_amount = 1.04
    
    def __init__(self, first, last, pay):
        self.first = first
        self.last = last
        self.pay = pay
        
        Empl.no_of_employees += 1


    def apply_raise(self):
        self.pay = int(self.pay * Empl.raise_amount)
```

# Inheritance

```python
class Parent:
    def greet(self):
        print("Hello from parent")

class Child(Parent): # Child class inheriting the Parent class
    def greet(self):
        super().greet()# calling greet function from parent class
        print('Hello from Child')

obj = Child() # Creating and object for Child class
obj.greet() # Accessing the inherited function from the parent class
```

```python
class Inherit_Empl(Empl):
    pass

obj1 = Inherit_Empl('Kartikay','Sharma',40000)
```

# Multiple Inheritance 

```python
class A:
    def method_a(self):
        print("Method A")

class B:
    def method_b(self):
        print("Method B")

class C(A,B): # Class C inherits from class A & B
    pass      # no need to create object inside the class

# creating an object after the class is fully defined
obj = C()

obj.method_a()
obj.method_b()
        
```

```python
# Multi Level Inheritance

class D:
    def method_a(self):
        print ("Method A")
class E(D):
    def method_b(self):
        print ("Method B")
class F(E):
    pass

obj = F()
obj.method_a()
obj.method_b()
```

```python
class Contact:
    all_contacts = []

def __init__(self, name, email):
    self.name = name
    self.email = email
    Contact.all_contacts.append(self)
```

```python
class Supplier(Contact):
    def order(self, order):
        print("send" "{} order to {}".format(order, self.name))
```

```python
c = Contact("somebody", "somebody@gmail.com")
s = Supplier("Suplier", "suplier@gmail.com")

print(c.name, c.email, s.name, s.email)
```

```python
# Method Overriding and Super()
class Parent:
    def greet(self) :
        print ('Hello from parent')
class Child(Parent):
    def greet(self) :
        super().greet()
        print( 'Hello from child')
obj1 = Child()
obj1.greet()
```
