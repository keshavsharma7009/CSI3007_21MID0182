```python
# Passing arguments-: Positional Argument

def describe_pet(animal_type, pet_name):
    """Display information about a pet"""
    print(f"\nI have a {animal_type}.")
    print(f"My {animal_type}'s name is {pet_name.title()}.")

describe_pet('dog', 'max')
```

```python
# Passing Arguments-: Keyword Argument

def describe_pet(animal_type, pet_name):
    """Display information about a pet"""
    print(f"\nI have a {animal_type}.")
    print(f"My {animal_type}'s name is {pet_name.title()}.")

name = input("Enter the pet name: ") 
breed = input("Enter animal type: ")
describe_pet(animal_type=breed, pet_name=name)
```

```python
#Multiple Function Calls

def say_hello(name):
    print(f"Hi {name}!")

say_hello("Alice")
say_hello("Bob")
say_hello("Charlie")
```

```python
# Order Matters in Positional Arguments

def introduce(name, city):
    print(f"{name} lives in {city}.")

introduce("Ravi", "Delhi")   # Correct ordwr
introduce("Delhi", "Ravi")   # Wrong meaning and order
```

```python
# Keyword Arguments
def show_info(name, subject):
    print(f"{name} is studying {subject}.")

show_info(name="Anu", subject="Math")
show_info(subject="Science", name="Keshav")
```

```python
# Default Values in arguments

def describe_pet(pet_name, animal_type="dog"):
    print(f"I have a {animal_type} named {pet_name}.")

describe_pet("Bruno")  # default = dog
describe_pet("Tommy", "cat")
```

```python
def describe_pet(pet_name, animal_type="dog"):
    print(f"I have a {animal_type} named {pet_name}.")

# Positional, uses default animal_type
describe_pet("bruno")

# Keyword argument, uses default animal_type
describe_pet(pet_name="bruno")

# Positional for both arguments
describe_pet("harry", "hamster")

# Keyword arguments
describe_pet(pet_name="harry", animal_type="hamster")

# Order doesn’t matter in keyword arguments
describe_pet(animal_type="hamster", pet_name="harry")
```

```python
# Making an argument optional

def greet_user(name, middle_name=""):
    if middle_name:
        print(f"Hello {name} {middle_name}")
    else:
        print(f"Hello {name}")

greet_user("Arun")
greet_user("Arun", "Kumar")
```

```python
# Returning a dictionary using a function

def build_person(name, age):
    return {"name": name, "age": age}

person = build_person("Nina", 22)
print(person)
```

```python
# passing a list as an argument
def greet_all(names):
    for name in names:
        print(f"Hello {name}")

names_list = input("Enter names separated by commas: ").split(",")
greet_all([name.strip() for name in names_list])
```

```python
def myfunc(name, lastname):
    for n in name:
        for m in lastname:
            print(f'Hi my name is {n} {m}.')

myfunc(name=['keshav','Jaish','Shriharsh'],lastname=['sharma','arun','kotecha'])
```

```python
# Passing Arbitary number of arguments

def make_pizza(*toppings):
    print("Toppings selected:")
    for t in toppings:
        print(f"-: {t}")

make_pizza("cheese", "tomato", "olives")
```

```python
# Using both positional and keyword arguments

def order_pizza(size, *toppings):
    print(f"\nMaking a {size} pizza with:")
    for t in toppings:
        print(f"- {t}")

order_pizza("medium", "corn", "mushroom", "extra cheese") # the data is entered in postional arg style but the *toppings if keyword arg style
```

```python
# using arbitary keyword arguments

def build_profile(**info):
    for key, value in info.items():
        print(f"{key}: {value}")

build_profile(name="Ravi", age=30, city="Mumbai")
```

```python
# Modifying a list in a function
# Function that modifies the list
def add_item(my_list):
    new_item = input("Enter an item to add to the list: ")
    my_list.append(new_item)
    print("List after modification inside the function:", my_list)

# Original list
items = ["apple", "banana"]
print("Original list before function call:", items)

add_item(items)

print("Original list after function call:", items)
```

```python
# Protecting a list from updates

# Method-1 converting the list into a tuple-:

# Function tries to modify the data, but it’s a tuple now
def try_to_modify(data):
    try:
        data.append("new_item")
    except AttributeError:
        print("This list is protected! You can't modify a tuple.")

# Original list turned into a tuple
original_list = ["dog", "cat"]
protected_list = tuple(original_list)# We typecasted the list into a tuple

print("Original (protected) list:", protected_list)
try_to_modify(protected_list)# Calling the function to check if we can still add elements to the list 
```

```python
# Method-2 Using a copied list not an orignal list

# Function modifies the copy, not the original
def safe_modify(my_list):
    copied_list = my_list.copy()
    item = input("Enter an item to add (modifies only the copy): ")
    copied_list.append(item)
    print("Modified copy inside function:", copied_list)

# Original list remains unchanged
animals = ["lion", "tiger"]
print("Original list before function call:", animals)

safe_modify(animals)

print("Original list after function call:", animals)
```
