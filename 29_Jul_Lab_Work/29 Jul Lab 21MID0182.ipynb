```python

# Off-by-one behiaviour of list

my_list = input("Enter items separated by spaces: ").split()
print("\nYour list is:", my_list)
print("Length of list is:", len(my_list))

# Correct index: len - 1
print("Last item in the list:", my_list[len(my_list) - 1])
```

```python
#Slicing thorugh a list
a = [1,2,3,4,5,6,7,8,9]

print(a[:2])#First 2 elements

print(a[-2:])#last 2 elements

print(a[2:])# Third to last element
```

```python
my_fav = ['cake', 'pizza', 'stew']  # original

friend_fav1 = my_fav[:]     # makes a copy (different object)
friend_fav2 = my_fav        # makes a reference (same object)

print("Before modifying original:")
print("friend_fav1:", friend_fav1)
print("friend_fav2:", friend_fav2)

# Modify the original list by appending (not reassigning)
my_fav.append('curry')  # mutates the same list

print("\nAfter modifying original:")
print("friend_fav1:", friend_fav1)  # stays unchanged
print("friend_fav2:", friend_fav2)  # changes with my_fav
```

```python
# Looping through a list

colors = input("Enter some colors separated by spaces: ").split()

print("\nLooping through the list:")
for color in colors:
    print(color)
```

```python
# Looping through a sliced list
a = ['My','book','is','very','heavy']
for s in a[:3]:
    print(s.title()) # only works with text not with the integers
```

# Tuple

```python
t = (500,100)
print(t[1])
t[1] = -50
```

```python
#Looping thorugh a Tuple
t=(100,200,400,500)
for a in t:
    print(a)
```

```python
print(t[:3])# Slicing a Tuple
```

```python
# looping through a sliced tuple

places = input("Enter place names separated by spaces: ").split()
place_tuple = tuple(places)

print("\nOriginal tuple:", place_tuple)

# Slice first 3 places
sliced_tuple = place_tuple[:3]

print("Looping through sliced tuple:")
for place in sliced_tuple:
    print(place)
```
