# Writing Numeric Data into a CSV file

```python
# Adding Numeric data to financial_transaction.csv file 

# we are using numpy to add numeric data to the csv file above we used the pandas dataframe to do so
data = np.array([
    [1001, "2025-09-09", 250.75,],
    [1002, "2025-09-08", 100.00,],
    [1003, "2025-09-07", 50.50,],
    [1004, "2025-09-06", 500.00,]
])

# Save as CSV
np.savetxt(
    'financial_transaction.csv',
    data,
    delimiter=',',
    header='transaction_id,date,amount',# these are the column headings in the file
    comments='',
    fmt='%s'   # everything saved as string
)

print("Numeric data is written using numpy")
```

```python
import openpyxl
from openpyxl.chart import Reference, BarChart

# make a new excel workbook
wb = openpyxl.Workbook()

# get the first sheet
ws = wb.active
ws.title = "Sales Data"   # give sheet a name

# put some data in the sheet (sales for 2023 and 2024)
data = [
    ["Genre", "2023 Sales", "2024 Sales"],
    ["Action", 120, 150],
    ["Adventure", 80, 95],
    ["Sports", 150, 170],
    ["RPG", 90, 110],
    ["Racing", 60, 75],
    ["Shooter", 200, 230],
    ["Puzzle", 40, 60]
]

# write each row of data into the sheet
for row in data:
    ws.append(row)

# choose the numbers (sales) for the chart
values = Reference(
    ws,
    min_col=2,     # start from column 2 (sales)
    max_col=3,     # take till column 3 (two years)
    min_row=1,     # include header
    max_row=len(data)  # go till last row
)

# choose the names (genres) for the chart
cats = Reference(
    ws,
    min_col=1,     # column 1 has genres
    max_col=1,
    min_row=2,     # start after header
    max_row=len(data)
)

# make a column chart (vertical bars)
chart = BarChart()
chart.type = "col"
chart.style = 11   # give some style to the chart
chart.add_data(values, titles_from_data=True)  # add numbers
chart.set_categories(cats)                     # add genre names

# set titles for chart and axes
chart.title = "Total Sales by Genre (2023 vs 2024)"
chart.x_axis.title = "Genre"
chart.y_axis.title = "Units Sold"

# put chart on the sheet (starting at cell E5)
ws.add_chart(chart, "E5")

# save the excel file
wb.save("unique_sales_chart.xlsx")
print("Unique bar chart created and saved in 'unique_sales_chart.xlsx'")
```
