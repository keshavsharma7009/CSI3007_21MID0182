# Image Tutorial

```python
from PIL import Image

import matplotlib.pyplot as plt
import numpy as np
```

```python
img = np.asarray(Image.open('stinkbug.png'))# Import an image and convert it into a NumPy array
print(repr(img))# Print the array in a way that shows all details (representation of the array)
```

```python
imgplot = plt.imshow(img)
```

```python
# Take only the first color channel (Red channel) from the image
# img[:,:,0] means: take all rows (:), all columns (:), and only channel 0
lum_img = img[:, :, 0]

# Show this single channel image using matplotlib
plt.imshow(lum_img)
```

```python
print(lum_img)
```

```python
print(repr(lum_img))
```

```python
plt.imshow(lum_img, cmap="hot")# Show the red channel of the image but use the "hot" colormap
# "hot" makes dark areas look black and bright areas look yellow/red/white
```

```python
# Show the red channel image using matplotlib
imgplot = plt.imshow(lum_img)
# Change the color map to 'nipy_spectral'
# This color map shows the image with many different colors like a rainbow
imgplot.set_cmap('nipy_spectral')
```

```python
# Show the red channel of the image
imgplot = plt.imshow(lum_img
# Add a color bar on the side
# The color bar shows the scale of values (dark to bright) used in the image
plt.colorbar()
```

```python
# Make a histogram of all pixel values from the red channel image
# lum_img.ravel()  flattens the 2D image into a 1D list of numbers
# bins=range(256)  groups values into 256 bins (0255 pixel intensity)
# fc='k', ec='k'  fill color (black) and edge color (black) for the bars
plt.hist(lum_img.ravel(), bins=range(256), fc='k', ec='k')
```

```python
print(repr(lum_img))
```

```python
# Show the red channel image but set limits for brightness display
# clim=(0, 175) means:
#   - Treat 0 as the darkest value
#   - Treat 175 as the brightest value
#   - Anything brighter than 175 will also look white
plt.imshow(lum_img, clim=(0, 175))
```

```python
imgplot = plt.imshow(lum_img)
imgplot.set_clim(0, 175)
```

```python
img = Image.open('stinkbug.png')
img.thumbnail((64, 64))  # resizes image in-place
# thumbnail() changes the original image (in-place)
imgplot = plt.imshow(img)
```

```python
imgplot = plt.imshow(img, interpolation="bilinear")
# Show the image using matplotlib
# interpolation="bilinear" smooths the image when it is resized
```

```python
imgplot = plt.imshow(img, interpolation="bicubic")
# Show the image using matplotlib
# interpolation="bicubic" makes the resized image even smoother than "bilinear"
```
