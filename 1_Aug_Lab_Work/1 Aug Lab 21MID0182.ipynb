# Sequence Unpacking

```python
odds = [1,3,5,7]# Container holding the data 

a,b,c,d = odds #  Indivisualy assiging the data from the container to multiple variables
```

```python
print(a,b,c,d)
```

```python
a,b,*c = odds # Sequence number 2
```

```python
print(a,b,c) # Here the variable *c means that rest of the data elements will be assigned to variable c
```

```python
a,*b,c = [3,5,7,9] # Sequence number 3
```

```python
print(a,b,c)# Here the var *b means that the data elements will be assigned to variable b and the last element will be for var c.
```

```python
# Sequence number 4

def a(x,y,z,p):
    return(x+y+z+p)
print(a(*odds))
```

```python
def say_hi(*name):
    for n in name:
        print(f"Hello, {n}")

say_hi("Kaushik", "Hari", "Darshini")
```

```python
# Sequence number 6 (Unpacking a dictionary)

def show_info(**info):
    for k, v in info.items():
        print(f"{k}  {v}")

course_info = {'code': 'CSI3009', 'name': 'python', 'venue': 'SJT317'}
show_info(**course_info)  # unpacking the dictionary
```

```python
# Creating an empty dictionary and then adding new values to it

my_dict = {}

print(my_dict)

my_dict[0] = 'Keshav'
my_dict[1] = 'Kavi'

print(my_dict)
    
```
